```python
import pandas as pd
import joblib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge

# 1. Load and clean
df = pd.read_csv('./data/train.csv', encoding='latin1')
df = df.dropna(subset=['text', 'sentiment'])

# 2. Map sentiments
sentiment_mapping = {'negative': -1.0, 'neutral': 0.0, 'positive': 1.0}
df['target_score'] = df['sentiment'].map(sentiment_mapping)

# 3. Define Inputs - REMOVED 'Time of Tweet'
X = df[['text', 'Age of User']]
y = df['target_score']

# 4. Build Pipeline - REMOVED 'Time of Tweet' from the encoder
preprocessor = ColumnTransformer(
    transformers=[
        ('text', TfidfVectorizer(max_features=2000), 'text'),
        ('cat', OneHotEncoder(handle_unknown='ignore'), ['Age of User'])
    ])

model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', Ridge(alpha=1.0))
])

# 5. Train and Export
print("Training new model...")
model_pipeline.fit(X, y)
joblib.dump(model_pipeline, 'sentiment_pipeline.joblib')
print("Model successfully updated and exported!")
```
