```python
import pandas as pd
import joblib
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge

# 1. Load Data (Fixing the encoding issue here!)
# Using 'latin1' handles the 'Â²' symbol perfectly.
df = pd.read_csv('./data/train.csv', encoding='latin1')

# 2. Clean Data
# Drop rows where 'text' or 'sentiment' might be missing
df = df.dropna(subset=['text', 'sentiment'])

# 3. Map sentiments to your desired scale (-1.0 to 1.0)
sentiment_mapping = {
    'negative': -1.0,
    'neutral': 0.0,
    'positive': 1.0
}
df['target_score'] = df['sentiment'].map(sentiment_mapping)

# 4. Define our Inputs (X) and Output (y)
# Let's use the text, plus two extra features that are easy for a user to provide
X = df[['text', 'Time of Tweet', 'Age of User']]
y = df['target_score']

# 5. Build a Preprocessing Pipeline
# This handles the text vectorization AND the categorical features automatically!
preprocessor = ColumnTransformer(
    transformers=[
        ('text', TfidfVectorizer(max_features=2000), 'text'),
        ('cat', OneHotEncoder(handle_unknown='ignore'), ['Time of Tweet', 'Age of User'])
    ])

# 6. Build the Final Model Pipeline
# We use Ridge Regression because it naturally outputs a continuous number 
# rather than just strict categories.
model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', Ridge(alpha=1.0))
])

# 7. Train the Model
print("Training model...")
model_pipeline.fit(X, y)
print("Training complete!")

# 8. Export the ENTIRE pipeline
# Because we used a Pipeline, this single file contains your vectorizer, 
# your encoders, AND the trained model!
joblib.dump(model_pipeline, 'sentiment_pipeline.joblib')
print("Model exported as 'sentiment_pipeline.joblib'")
```
